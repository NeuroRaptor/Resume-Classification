```python
import os
import re
import glob
import win32com.client

# Path to the new folder containing .doc files
input_folder = "C:/Users/thila/OneDrive/Desktop/ML Project UE/Resumes_Dataset"
file_pattern = os.path.join(input_folder, "*.doc")

# Initialize Word application
word = win32com.client.Dispatch("Word.Application")
word.Visible = False  # Keep the Word application invisible

try:
    for i, doc in enumerate(glob.iglob(file_pattern)):
        try:
            # Get the absolute path of the .doc file
            in_file = os.path.abspath(doc)

            # Open the .doc file
            wb = word.Documents.Open(in_file)

            # Create the new .docx filename
            out_file = re.sub(r'\.\w+$', '.docx', in_file)

            # Save as .docx format
            wb.SaveAs2(out_file, FileFormat=16)  # FileFormat=16 corresponds to .docx
            wb.Close()

            print(f"Converted: {in_file} -> {out_file}")
        except Exception as e:
            print(f"Error converting {doc}: {e}")
finally:
    # Quit Word application
    word.Quit()

print("Conversion process completed.")
```

```python
import os
import textract
import pandas as pd

# Directory containing resumes
directory1 = "C:/Users/thila/OneDrive/Desktop/Resumes/Peoplesoft"

# Lists to store resume text and categories
file_path1 = [] 
category1 = []

# Loop through all files in the directory
for i in os.listdir(directory1):
    if i.endswith('.docx'):
        file_path = os.path.join(directory1, i)
        try:
            # Extract text content from the .docx file
            text = textract.process(file_path).decode('utf-8')
            file_path1.append(text)
            category1.append('PeopleSoft')  # Assign category
        except Exception as e:
            print(f"Error processing file {file_path}: {e}")

# Create a DataFrame with the extracted data
data1 = pd.DataFrame({'Resume_Text': file_path1, 'Category1': category1})

# Display the resulting DataFrame
print(data1)

```
